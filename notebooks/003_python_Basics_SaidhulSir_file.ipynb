### String Indexing
***
- `String indexing` in Python allows you to access individual characters within a string by their position.

- Python uses `zero-based indexing`, which means that the index of the first character in a string is 0, the index of the second character is 1, and so on.

- `Negative indices` are also supported, allowing you to count positions from the end of the string, where -1 refers to the last character, -2 refers to the second-to-last character, and so forth.

<img src="images/python_string.png" alt="Concept-Alert" style="width: 300px;float:center; margin-right:15px" width="300" height="100"/>
<br />



```python
# Define a string
my_string = "Shaik Saidhul"

# Accessing individual characters using positive indexing
first_character = my_string[0]  # 's'
second_character = my_string[1] # 'h'

# Accessing individual characters using negative indexing
last_character = my_string[-1]   # 'l'
second_last_character = my_string[-2]  # 'u'

# Printing the characters
print("First character:", first_character)
print("Second character:", second_character)
print("Last character:", last_character)
print("Second last character:", second_last_character)
```

## Python Operators

In Python, operators are symbols that perform operations on operands, such as variables, values, or expressions. Python supports various types of operators, including:

#### 1. Arithmetic Operators:
Arithmetic operators are used to perform mathematical operations like addition, subtraction, multiplication, division, etc.
- Addition: `+`
- Subtraction: `-`
- Multiplication: `*`
- Division: `/`
- Modulus (Remainder): `%`
- Exponentiation: `**`
- Floor Division (Integer Division): `//`

#### 2. Comparison Operators:
Comparison operators are used to compare the values of two operands and return a Boolean value.
- Equal to: `==`
- Not equal to: `!=`
- Greater than: `>`
- Less than: `<`
- Greater than or equal to: `>=`
- Less than or equal to: `<=`

#### 3. Logical Operators:
Logical operators are used to combine multiple conditions or expressions and return a Boolean value.
- AND: `and`
- OR: `or`
- NOT: `not`

#### 4. Assignment Operators:
Assignment operators are used to assign values to variables.
- Assignment: `=`
- Addition Assignment: `+=`
- Subtraction Assignment: `-=`
- Multiplication Assignment: `*=`
- Division Assignment: `/=`
- Modulus Assignment: `%=`
- Exponentiation Assignment: `**=`
- Floor Division Assignment: `//=`

#### 5. Membership Operators:
Membership operators are used to test if a value is present in a sequence.
- `in`: Returns `True` if the value is present in the sequence.
- `not in`: Returns `True` if the value is not present in the sequence.

# Sets

 - a set is an `unordered` collection of `unique` elements.
   
 - Sets are `mutable`, meaning you can add or remove elements from them, but the elements themselves are immutable.

 - Sets are often used for  Membership Testing, removing duplicates from a sequence, and performing mathematical operations such as intersection, union, difference, and symmetric difference.
   
 - Sets are created using curly braces `{}` or the `set()` constructor.

### ``Creating a Set:``

```python
my_set = {1, 2, 3}
empty_set = set()
print(type(my_set))
```

### `Adding and Removing Elements:`

```python
my_set = {1, 2, 3}

# Adding elements
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}

# Removing elements
my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}

# Removing elements safely
my_set.discard(5)  # No error if element doesn't exist
print(my_set)      # Output: {1, 3, 4}
```

### `Set Operations:`

    | => union
    & => intersection
    - => difference
    ^ => symm difference

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Union
union_set = set1 | set2
print(union_set)  # Output: {1, 2, 3, 4, 5}

# Intersection
intersection_set = set1 & set2
print(intersection_set)  # Output: {3}

# Difference
difference_set = set1 - set2
print(difference_set)  # Output: {1, 2}

# Symmetric Difference
symmetric_difference_set = set1 ^ set2
print(symmetric_difference_set)  # Output: {1, 2, 4, 5}
```

### `Set Membership Testing:`

```python
my_set = {1, 2, 3}

# Membership testing
print(1 in my_set)  # Output: True
print(4 in my_set)  # Output: False
```

```python
# remove a duplicates from the list

l1 = [1,2,2,3,3,3,4,4,4,4,5,5]

list(set(l1))
```

# If,elif,else Statements

 - `if`, `elif`, and `else` are control flow statements used for conditional execution.

 - They allow you to execute different blocks of code based on specified conditions.


        if condition1:
            perform action1
        elif condition2:
            perform action2
        elif condition3:
            perform action3
        else:

# For Loop
***

- `for loop` is used to iterate over a sequence (such as a list, tuple, string, or range) or any iterable object.
        
- It allows you to execute a block of code repeatedly for each `item` in the `sequence`.

Here's the general format for a **for** loop in Python:

    for item in sequence:
        # Code block to be executed for each item

1. The `item` variable is used to represent each item in the sequence during each iteration.
2. The `sequence` can be any iterable object like a `list, tuple, string, range, etc`.
3. The code block indented under the for statement is executed for each item in the sequence.

#### `Example 1: Iterating over a List:`

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

#### `Example 2: Iterating over a String:`

```python
for char in "hello":
    print(char)
```

#### `Example 3: Using Range:`

```python
for i in range(5):
    print(i)
```

#### `Example 4: Nested for Loops:`

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        print(x, y)
```

```python
my_dict = {"name": "John", "age": 30, "city": "New York"}

for key, value in my_dict.items():
    if "John" in my_dict.values():
        print(key, ":", value)
    else:
        print("not found")
```

# While loops
***

- `while loop` is used to repeatedly execute a block of code as long as a specified condition is true.
- It continues iterating until the condition becomes false.

The general format of a while loop is:

    while condition:
        # Code block to be executed as long as the condition is true

- The condition is a boolean expression that is evaluated before each iteration of the loop.
- The code block inside the loop is executed repeatedly as long as the condition remains true.

```python
# Printing Numbers from 1 to 5
num = 1
while num <= 5:
    print(num)
    num += 1
```

```python
# Sum of Numbers from 1 to 100

i = 1
sum = 0

while i <= 100:
    sum += i
    i += 1
print(sum)
```

```python
# check user input is correct password

password = "python@123"
user_input = ""

while user_input != password:
    user_input = input("Enter the password: ")
print("Access granted!")
```

```python
# Using while loop with else block

num = 1
while num <= 5:
    print(num)
    num += 1
else:
    print("Loop executed successfully!")
```

```python
# check user input is correct password

password = "python@123"
user_input = ""
attempt = 0
max_attempts = 3

while user_input != password and attempt < max_attempts:
    user_input = input("enter password: ")
    attempt += 1
    if user_input == password:
        print("access granted")
    elif attempt == max_attempts:
        print("wrong password too many attemps, reset password")
```

## Exception Handling
***

    Exception handling in Python is a mechanism that allows you to deal with errors or exceptional situations that may occur during the execution of your code. By handling exceptions, you can gracefully recover from errors and prevent your program from crashing.

#### try and except

- The try block is used to enclose the code that may raise an exception.
- The except block is used to handle specific types of exceptions that occur within the try block.

```python
# a = int(input("a = "))
# b = int(input("b = "))

# result = a/b
# print(result)
```

```python
try:
    a = int(input("a = "))
    b = int(input("b = "))

    result = a/b
    print(result)

except ZeroDivisionError:
    print("Cannot divide by zero!")
```

#### `Handling Multiple Exceptions:`

```python
try:
    for i in [1,2,3]:
        a = int(input("a = : "))
        b = int(input("b = : "))

        result = a/b
        print(result)

except ZeroDivisionError:
    print("Cannot divide by zero!")
except (ValueError,TypeError,NameError):
    print("Invalid operation or data type or not available")
except Exception as e:
    print(f"please enter inputs propery, check for error below : \n {e}")
```

```python
# letter count in a string

s = "sjsSS"
s = s.lower()

def letter_count(input):

    empty_dict = {}

    for i in input:
        if i not in empty_dict:
            empty_dict[i] = 1
        else:
            empty_dict[i] += 1

    return empty_dict

letter_count(s)
```

```python
# word count in a sentence

sentence = "This is is a a a test sentence"

def word_count(input):

    empty_dict = {}
    words = input.split(" ")

    for i in words:
        if i not in empty_dict:
            empty_dict[i] = 1
        else:
            empty_dict[i] += 1

    return empty_dict

print("Number of words:", word_count(sentence))
```

## Lambda functions

- are also known as anonymous functions, are small, inline functions that can have any number of arguments but only one expression.

- They are defined using the lambda keyword and are useful for creating short, throwaway functions without needing to define a separate named function.

- Here's the basic syntax of a lambda function:

- `lambda arguments: expression`


```python
result = lambda a,b,c : (a+b+c)/3

result(3,4,5)
```

```python
# python function to reverse a string
def reverse(input):
    return input[::-1]

reverse("shaik")
```

```python
# # python lambda function to reverse a string

reverse = lambda input: input[::-1]

reverse("shaik")
```
